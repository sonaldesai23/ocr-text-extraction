Converted PDFs into images using pdf2image and applied Tesseract OCR for text recognition.

```python
import pytesseract
from pdf2image import convert_from_path
from PIL import Image
from google.colab import files
```

```python
# Set pytesseract path (Colab usually detects automatically)
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'
```

```python
# -------------------------------
# 1. Extract text from an IMAGE
# -------------------------------
print("ðŸ“Œ Please choose an image file (JPG/PNG)...")
uploaded_img = files.upload()  # Opens a file chooser dialog
image_path = list(uploaded_img.keys())[0]  # Get file name
```

```python
# Open and process the image
img = Image.open(image_path)
text_from_image = pytesseract.image_to_string(img, lang='eng')
```

```python
print("\nâœ… Extracted Text from Image:\n")
print(text_from_image)
```

```python
# -------------------------------
# 2. Extract text from a PDF
# -------------------------------
print("\nðŸ“Œ Please choose a PDF file...")
uploaded_pdf = files.upload()  # Opens a file chooser dialog
pdf_path = list(uploaded_pdf.keys())[0]
```

```python
# Convert PDF pages to images
pages = convert_from_path(pdf_path, dpi=300)
```

```python
pdf_text = ""
for page_num, page in enumerate(pages):
    text = pytesseract.image_to_string(page, lang='eng')
    pdf_text += f"\n--- Page {page_num+1} ---\n{text}"

print("\nâœ… Extracted Text from PDF:\n")
print(pdf_text)
```
